## Calculating Moth Echolocation

The code below is a solution to the following problem:

The Nightly War of Moths and Bats. Moths, like any other animal, do not like to be eaten.
Unfortunately for many of them, they are being hunted by bats at night. 
The bats use ultrasound echolocation to detect moths in light and catch them.
Over the course of evolution, many moth species have developed
very simple ears that essentially act as bat detectors - they have one
or two auditory receptor neurons that are only sensitive to high
frequency sounds. The moth's response to an incoming bat is
simple: if the bat is still far away, it will simply y away from the
sound. If the bat is relatively close, the moth will fold its wings and
let itself drop onto the ground (see for instance: Roeder KD, 1963,
Nerve Cells and Insect Behavior, Harvard University Press).

We want to think about how the moth retrieves the relevant information (distance and incoming
angle of the bat) from the sound intensity it picks up on its left and right ear. 
Given a bat coming from angle f at distance r , we assume that the sound intensities 
impinging on the moths two ear are given by:

    I right = I/r(cos^2(f/2) + α)
    I left = I/r(sin^2(f/2) + α)

We ignore the problem of how the moth distinguishes between sounds coming from the front or
the back and will simply focus on the front (f in [0; 180]).

1. Sketch the intensities on the left and right ear against the angle of the incoming bat for a fixed distance.
2. Sketch the intensities IL and IR against the distance of the incoming bat for a fixed angle.
3. What computations does the moth have to perform on IL and IR to extract the angle f of the
    incoming bat and its distance r? What does the neuron compute that decides whether the moth
    should let itself fall to the ground?

```python
# Let’s assume that I = 1, r = 1 and α = 0. Given that the angle is between 0 and 180
# we have the following distribution of sound intensities:

# Import libraries
import matplotlib.ticker as tck
import matplotlib.pyplot as plt 
import numpy as np 
  
# Creating vectors X and Y 
x = np.linspace(0, np.pi, 100) 
y_l = np.sin(x/2) ** 2
y_r = np.cos(x/2) ** 2

fig, ax = plt.subplots(figsize = (10, 5)) 

# Create the plot 
plt.plot(x, y_l, 'c', label='Left receptor') 
plt.plot(x, y_r, 'm', label='Right receptor')

plt.legend(loc='upper center')

ax.set_xlabel('Angle ($\phi$)')
ax.set_ylabel('Bat sound intensity (I)')

tick_pos= [0, np.pi/2 , np.pi]
labels = ['0', '$\pi/2$', '$\pi$']
plt.xticks(tick_pos, labels)

# Show the plot 
plt.show()
```

```python
# Let’s assume that I = 1, φ = π/3, and α = 0. 
# Given that we have the following distribution of sound intensities against the distance. 
# Such a picture will hold for all the angles except for π/2,
# where the intensities for both left and right auditory receptors are equal.

# Creating vectors X and Y 
x = np.linspace(0, 3, 100) 
y_l = 3/(4*x)
y_r = 1/(4*x)

fig, ax = plt.subplots(figsize = (10, 5)) 

# Create the plot 
plt.plot(x, y_l, 'c', label='Left receptor') 
plt.plot(x, y_r, 'm', label='Right receptor')

plt.legend(loc='upper center')

ax.set_xlabel('Distance (r)')
ax.set_ylabel('Bat sound intensity (I)')

#tick_pos= [0, np.pi/2 , np.pi]
#labels = ['0', '$\pi/2$', '$\pi$']
#plt.xticks(tick_pos, labels)

# Show the plot 

plt.show()
```
